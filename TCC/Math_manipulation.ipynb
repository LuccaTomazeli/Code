```python
from sympy import symbols, exp, log, plot
```

```python
# Define variables
pH, pH_ideal, pH_ideal_range = symbols('H pH_ideal pH_ideal_range')

# Define the function
#ph = -log(10, H)
f = exp(-((pH - pH_ideal) / pH_ideal_range)**2)

# Substitute specific values for pH_ideal and pH_ideal_range for plotting
f_sub = f.subs({pH_ideal: 4.25, pH_ideal_range: 2})

# Plot the function for H in a suitable range (e.g., 0.0001 to 10)
plot(f_sub, (pH, 0.001, 14), title="Plot of the function",
     xlabel="pH", ylabel="mu_pH");
```
